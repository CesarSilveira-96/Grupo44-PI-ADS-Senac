```python
import pandas as pd

df = pd.read_csv('petrol_prices_comparison.csv')

df.head()
```

```python
df.info()
```

```python
df = df.drop_duplicates()
df = df.dropna()

df.head()
```

```python
df['Price_Diff'] = df['Mar7_USD'] - df['Before_War_USD']

df.head()
```

```python
def impacto(x):
    if x > 15:
        return 'Alto'
    elif x > 5:
        return 'Médio'
    else:
        return 'Baixo'

df['Impact_Level'] = df['Pct_Increase'].apply(impacto)

df.head()
```

```python
df.to_csv('dados_tratados.csv', index=False)
```

```python
df_oil = pd.read_csv('crude_oil_daily.csv')

df_oil.head()
```

```python
df_oil.info()
```

```python
df_oil['Date'] = pd.to_datetime(df_oil['Date'])

df_oil.head()
```

```python
df_oil['Avg_Price'] = (df_oil['Brent_USD'] + df_oil['WTI_USD']) / 2

df_oil.head()
```
